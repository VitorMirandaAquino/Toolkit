# Resumo: Técnicas de Pré-processamento de Texto e Vetorização
Este notebook tem como objetivo fornecer uma compreensão clara e detalhada das principais etapas de pré-processamento de texto e técnicas de vetorização, essenciais para o processamento de linguagem natural (NLP). O conteúdo abrange as seguintes áreas:

## Pré-processamento de Texto
1. Limpeza de Dados: Descrevemos métodos para transformar o texto em um formato uniforme e mais fácil de manipular. Isso inclui a conversão para minúsculas e remoção de acentos e pontuações.

2. Remoção de Stopwords: Abordamos a eliminação de palavras frequentes, mas não informativas, que podem prejudicar a análise e a modelagem.

3. Stemming e Lemmatization: Explicamos as técnicas de stemming, que reduzem as palavras à sua forma básica, e lemmatization, que as reduz a suas formas de dicionário, melhorando a eficácia do processamento.

## Tokenização
1. Espaço em Branco: Mostramos como dividir o texto em tokens com base em espaços em branco, usando a função WhitespaceTokenizer do NLTK.

2. Pontuação: Demonstramos a tokenização com base em pontuação, utilizando a função WordPunctTokenizer do NLTK para lidar com a separação de palavras e sinais de pontuação.

## Vetorização
1. Count Vectorizer (Bag of Words): Exploramos o uso do Count Vectorizer para converter o texto em uma matriz numérica, representando a frequência das palavras.

2. TF-IDF Vectorizer: Introduzimos o TF-IDF Vectorizer, que leva em consideração a importância das palavras em relação ao documento e à coleção, fornecendo uma representação mais significativa.

## Visualização de Palavras Frequentes
Por fim, demonstramos como visualizar as palavras mais frequentes em um corpus de texto, oferecendo insights valiosos para análises posteriores.

## 4.2. TF-IDF
O TF-IDF (Frequência do Termo-Inverso da Frequência nos Documentos) é uma técnica de processamento de linguagem natural. Ela avalia a importância de palavras em um documento em relação a uma coleção de documentos. Palavras frequentes no documento, mas raras na coleção, recebem maior destaque. Isso é alcançado calculando a Frequência do Termo (TF), que mede quantas vezes uma palavra aparece no documento, e a Frequência Inversa nos Documentos (IDF), que mede a raridade da palavra. O valor final do TF-IDF é o produto do TF e IDF. A técnica é útil para classificação de documentos, recuperação de informações e análise de texto, convertendo palavras em representações numéricas para algoritmos de aprendizado de máquina.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

```

```python
df_texto['texto_tratado'] = df_texto['texto'].apply(nlp_functions.remove_stopwords_portuguese).apply(nlp_functions.remove_punctuation)
```

```python
def nuvem_palavras(df, coluna):
    # Concatena todos os textos em uma só string
    todas_palavras = ' '.join([texto for texto in df[coluna]])

    # Cria a nuvem de palavras com os parâmetros
    nuvem_palavras = WordCloud(width=800, height=500, max_font_size=110,
                               collocations=False).generate(todas_palavras)
    
    # Plota a imagem
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
nuvem_palavras(df_texto, 'texto_tratado')
```

## 5.2. Pareto

```python
import seaborn as sns
```

```python
def pareto(df, coluna, quantidade):
    # Concatena todas os textos em uma só string
    todas_palavras = ' '.join([texto for texto in df[coluna]])
    # Cria um dicionário com as palavras e a contagem das frequências
    frequencia = nltk.FreqDist(nlp_functions.tokenize_whitespace(todas_palavras))
    # Cria um dataframe desse dicionário
    df_frequencia = pd.DataFrame({"Palavra": list(frequencia.keys()),
                                  "Frequência": list(frequencia.values())})
    # Filtra o dataframe para a quantidade de palavras desejadas no gráfico
    df_frequencia = df_frequencia.nlargest(columns= "Frequência", n=quantidade)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequencia, x= "Palavra", y = "Frequência", color = 'blue')
    ax.set(ylabel = "Contagem")
    plt.show()   
```

```python
pareto(df_texto, 'texto_tratado', 10)
```

# 6. Classificador de texto

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
def classificar_texto(df, coluna_texto, coluna_classificacao):
    # Instancia o vetorizador com um número máximo de features 
    vetorizar = CountVectorizer(lowercase=False, max_features=50)
    # Vetoriza o corpus fornecido
    bag_of_words = vetorizar.fit_transform(df[coluna_texto])
    # Separa os dados em treino em teste e treino
    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                                 df[coluna_classificacao],
                                                                 random_state = 42)
    
    # Instancia o modelo de Regressão Logistíca
    regressao_logistica = LogisticRegression(solver="lbfgs")
    # Treina o modelo
    regressao_logistica.fit(treino, classe_treino)
    # Retorna a acurácia do teste
    return regressao_logistica.score(teste, classe_teste)
    
```
